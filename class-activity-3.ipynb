```python
import pandas as pd
```

```python
errorsraw = pd.read_csv('excelErrorsRaw.csv', dtype='str')
errorspost = pd.read_csv('excelErrorsPost.csv', dtype='str')
```

```python
errorsraw
```

```python
errorspost
```

```python
changed_samples = []
updated_samples = []
change_count = 0
for i in range(0,99):
    if(errorsraw.iloc[i,0] != errorspost.iloc[i,0]):
        changed_samples.append(errorsraw.iloc[i,0])
        updated_samples.append(errorspost.iloc[i,0])
        change_count += 1
print(change_count)
print(changed_samples)
print(updated_samples)
```
